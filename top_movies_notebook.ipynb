```python
import os
import re
from collections import defaultdict

# Directory containing reviews
train_dir = 'aclImdb/train/'

# Dictionary to store the sum of ratings and the number of reviews per movie
movie_ratings = defaultdict(list)

# Function to process reviews
def process_reviews(directory):
    for sentiment in ['pos', 'neg']:
        sentiment_dir = os.path.join(directory, sentiment)
        for filename in os.listdir(sentiment_dir):
            if filename.endswith('.txt'):
                # Extract ID and rating from file name
                match = re.match(r'(\d+)_(\d+)\.txt', filename)
                if match:
                    movie_id = match.group(1)
                    rating = int(match.group(2))
                    # Add the rating to the corresponding movie list
                    movie_ratings[movie_id].append(rating)

# Processing reviews in the train directory
process_reviews(train_dir)

# Calculate the average rating per movie
movie_avg_ratings = {movie_id: sum(ratings) / len(ratings) for movie_id, ratings in movie_ratings.items()}

# Sort by rating (lowest to highest) and then by ID (lowest to highest)
sorted_movies = sorted(movie_avg_ratings.items(), key=lambda x: (x[1], x[0]))

# Get the top 10 worst rated movies
worst_movies = sorted_movies[:10]

# Get the top 10 best rated movies
best_movies = sorted_movies[-10:]

# Print the results
print("Top 10 worst rated movies on average:")
for movie_id, avg_rating in worst_movies:
    print(f"Movie ID: {movie_id}, Average Rating: {avg_rating:.2f}")

print("\nTop 10 best rated movies on average:")
for movie_id, avg_rating in best_movies:
    print(f"Movie ID: {movie_id}, Average Rating: {avg_rating:.2f}")
```
